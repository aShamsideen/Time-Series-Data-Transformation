### Power Transform

```python
import pandas as pd
from scipy.stats import boxcox
```

```python
births = pd.read_csv('daily-total-female-births.csv')
```

```python
data = births['Births']
result, lmbda = boxcox(data)
```

```python
from math import log
from math import exp

def invert_boxcox(value, lam):
    if lam == 0:
        return exp(value)
    return exp(log(lam * value + 1) / lam)
```

```python
datas = data
datas = [x for x in datas]
print(datas)
```

```python
transformed, lmbda = boxcox(datas)
print(transformed, lmbda)
```

```python
inverted = [invert_boxcox(x, lmbda) for x in datas]
print(inverted)
```

```python
from math import log
from math import exp
from scipy.stats import boxcox

def inverted_boxcox(value, lam):
    if lam  == 0:
        return exp(value)
    return exp(log(lam * value + 1) /lam)

data = [x for x in range(1, 10)]
print('Data:', data)

transformed, lmdda = boxcox(data)
print('Transformed:', transformed, lmbda)

inverted = [inverted_boxcox(x, lmbda) for x in transformed]
print('Invnerted:', inverted)
```

### Difference Transform

```python
def difference(data, interval):
    return [data[i] - data[i - interval] for i in range(interval, len(data))]

def invert_difference(orig_data, diff_data, interval):
    return [diff_data[i-interval] + orig_data[i-interval] for i in range(interval, len(orig_data))]

data = [x for x in range(1, 10)]
print('Data:', data)

transformed = difference(data, 1)
print('Transformed:', transformed)

inverted = invert_difference(data, transformed, 1)
print('Inverted:', inverted)
```

### Standardization

```python
from sklearn.preprocessing import StandardScaler
from numpy import array

data = [x for x in range(1, 10)]
data = array(data).reshape(len(data), 1)
print('Data:', data)

transformer = StandardScaler()
transformer.fit(data)

transformed = transformer.transform(data)
print('Transformed:', transformed)

inverse = transformer.inverse_transform(transformed)
print('Inverted:', inverted)
```

### Normalization

```python
from sklearn.preprocessing import MinMaxScaler
from numpy import array

data = [x for x in range(1, 10)]
data = array(data).reshape(len(data), 1)
print('Data:', data)

transformer = MinMaxScaler()
transformer.fit(data)

transformed = transformer.transform(data)
print('Transformed:', transformed)

inverted = transformer.inverse_transform(transformed)
print('Inverted:', inverted)
```
